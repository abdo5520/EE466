```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
bit_rate = 1  # Bits per second
carrier_frequency = 10  # Hz, common carrier frequency for ASK
sampling_rate = 100  # Hz, should be significantly higher than the carrier frequency
duration = 1 / bit_rate  # Seconds, duration of one bit
```

```python
# Binary data to modulate
data = np.array([0, 1, 0, 1, 1, 0, 1])
```

```python
# Generating time vector for one bit
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
```

```python
# Extended time vector for the entire data
time_vector = np.linspace(0, len(data) * duration, int(sampling_rate * duration * len(data)), endpoint=False)
```

```python
# Generating carrier signal for one period
carrier_signal = np.cos(2 * np.pi * carrier_frequency * t)
# Extended carrier signal for visualization
extended_carrier_signal = np.tile(carrier_signal, len(data))
```

```python
# Digital signal (original binary data repeated for the duration of one bit)
digital_signal = np.repeat(data, int(sampling_rate * duration))
```

```python
# ASK modulation
# In ASK, binary 1 is represented by the carrier signal, and binary 0 by a zero amplitude
modulated_signal = np.zeros(len(time_vector))
for i, bit in enumerate(data):
    if bit == 1:
        modulated_signal[i * len(t):(i+1) * len(t)] = carrier_signal
```

```python
# Plotting
plt.figure(figsize=(15, 8))

# Plot carrier signal before modulation
plt.subplot(3, 1, 1)
plt.plot(time_vector, extended_carrier_signal, label='Carrier Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Carrier Signal Before Modulation')
plt.grid(True)

# Plot original digital signal
plt.subplot(3, 1, 2)
plt.plot(time_vector, digital_signal, drawstyle='steps-post', label='Original Digital Signal')
plt.ylim(-0.5, 1.5)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Original Digital Signal')
plt.grid(True)

# Plot ASK modulated signal
plt.subplot(3, 1, 3)
plt.plot(time_vector, modulated_signal, label='ASK Modulated Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('ASK Modulated Signal')
plt.grid(True)

plt.tight_layout()
plt.show()
```
