```python
import numpy as np
import matplotlib.pyplot as plt
H = np.hstack((np.zeros(20), np.arange(10)/10, np.zeros(20)))
w = np.linspace(-0.5, 0.5, 50)   #parameters are start, end, number of samples in the H vector above
plt.plot(w, H, '.-')
plt.show()
```

```python
InversedH = (np.fft.ifft(np.fft.ifftshift(H)))
Shifted_InversedH = np.fft.ifftshift (InversedH)

plt.figure('impulse')
plt.plot(np.real(Shifted_InversedH), '.-')
plt.plot(np.imag(Shifted_InversedH), '.-')
plt.legend(['real','imag'], loc=1)
plt.show()
```

```python
Shifted_InversedH
```

```python
h = Shifted_InversedH
h_fft = np.fft.fftshift(np.abs(np.fft.fft(h, 1024)))
plt.figure('Frequency')
plt.plot(h_fft)
plt.show()
```

```python
window = np.hamming(len(h))
h = h * window

plt.figure('impulse')
plt.plot(np.real(h), '.-')
plt.plot(np.imag(h), '.-')
plt.legend(['real','imag'], loc=1)
plt.show()
```

```python
h_fft = np.fft.fftshift(np.abs(np.fft.fft(h, 1024)))
plt.figure('Frequency')
plt.plot(h_fft)
plt.show()
```
