```python
import numpy as np
import matplotlib.pyplot as plt

H = np.hstack((np.zeros(200), np.arange(100)/100, np.zeros(200)))
w = np.linspace(-0.5, 0.5, 500)
plt.plot(w, H, '.-')
plt.show()
```

```python
InversedH = (np.fft.ifft(np.fft.ifftshift(H)))
Shifted_InversedH = np.fft.ifftshift (InversedH)

# Now we plot the New Impulse Response 
plt.figure('impulse')
plt.plot(np.real(Shifted_InversedH), '.-')
plt.plot(np.imag(Shifted_InversedH), '.-')
plt.legend(['real','imag'], loc=1)
plt.show()
```

```python
h = Shifted_InversedH
h_fft = np.fft.fftshift(np.abs(np.fft.fft(h, 1024)))
plt.figure('Frequency')
plt.plot(h_fft)
plt.show()
```
